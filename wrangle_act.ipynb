### There are many fake names so we will correct them here

```python
def correct_name(row):    
    if row['name'] == 'a':            
        text = row['text']
        name_sub_list = text.split()
        
        name = 'None'

        if 'named' in name_sub_list:
            name_pos = name_sub_list.index('named') + 1
            name = name_sub_list[name_pos]
        
        elif 'name is' in text:            
            name_pos = name_sub_list.index('name') + 2
            name = name_sub_list[name_pos]
        
        row['name'] = name.replace('.','')
    
    return row
    
df_clean_twitter_archive = df_clean_twitter_archive.apply(correct_name, axis=1)
```

```python
# correcting the datatypes

df_clean_twitter_archive.timestamp = pd.to_datetime(df_clean_twitter_archive.timestamp)
df_clean_twitter_archive.retweeted_status_timestamp = pd.to_datetime(df_clean_twitter_archive.retweeted_status_timestamp)
```

```python
# Expanding url - in case it's not null

def expand_urls(row):    
    
    if pd.notnull(row['expanded_urls']):
        return row
    
    else:        
        tweet_id = row['tweet_id']        
        row['expanded_urls'] = 'https://twitter.com/dog_rates/status/{}'.format(tweet_id)
        
        return row

df_clean_twitter_archive = df_clean_twitter_archive.apply(expand_urls, axis=1)
```

```python
# Getting final rating in decimal form

def get_final_rating(row):
    #change it,if text matches with numerator
    decimal_numerator_pattern = '(\d+\.\d+)\/(\d+)'
    matches = re.search(decimal_numerator_pattern, row['text'])
    if matches != None:
        row['rating_numerator'] = matches.group(1)
        row['rating_denominator'] = matches.group(2)

    return row
        
df_clean_twitter_archive = df_clean_twitter_archive.apply(get_final_rating, axis=1)
```

```python
# Rectifying dog_stage

df_clean_final['dog_stage'] = 'None'

def dog_stage_acess(row):
    
    dog_stage = []
    
    if row['doggo'] == 'doggo':
        dog_stage.append('doggo')
    if row['floofer'] == 'floofer':
        dog_stage.append('floofer')
    if row['pupper'] == 'pupper':
        dog_stage.append('pupper')
    if row['puppo'] == 'puppo':
        dog_stage.append('puppo')
        
    if len(dog_stage) < 1: 
        row['dog_stage'] = 'None'
    else:         
        row['dog_stage'] = ','.join(dog_stage)
    
    return row

df_clean_final = df_clean_final.apply(dog_stage_acess, axis=1)

# Dropping redundant columns
df_clean_final = df_clean_final.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1)
```

### Favourites v/s Retweets
grouping them on the basis of dog stage

```python
df = df_clean_final[['retweet_count', 'favorite_count', 'dog_stage']]

y, x = df.retweet_count, df.favorite_count
mask = ~np.isnan(x) & ~np.isnan(y)
slope, intercept, r_value, p_value, std_err = stats.linregress(x[mask], y[mask])
line = slope*x+intercept
plt.xlabel('favorite_count')
plt.ylabel('retweet_count')
plt.title('Favorites v/s Retweets')
plt.plot(x, y, 'o', x, line, color='green',markerfacecolor = 'Black', linewidth = 1)

print('Line Equation: retweets = ({})*favorites + {}'.format(slope,intercept))
print('Correlation of line of best fit: {}'.format(r_value))

df_puppers = df[df['dog_stage'] == 'pupper']
df_puppo = df[df['dog_stage'] == 'puppo']
df_doggo = df[df['dog_stage'] == 'doggo']
df_floofer = df[df['dog_stage'] == 'floofer']
df_none = df[df['dog_stage'] == 'None']


bx = df_puppers.plot(kind = 'scatter', x='favorite_count', y='retweet_count', color='Red', alpha=0.5)
df_doggo.plot(kind = 'scatter', x='favorite_count', y='retweet_count', color='yellow', ax=bx, alpha=0.5)
df_puppo.plot(kind = 'scatter', x='favorite_count', y='retweet_count', color='Blue', ax=bx, alpha=0.5)
df_floofer.plot(kind = 'scatter', x='favorite_count', y='retweet_count', color='Green', ax=bx, alpha=0.5)


# Legend
puppers = mpatches.Patch(label='Pupper', color = 'Red')
doggo = mpatches.Patch(label='Doggo', color = 'yellow')
puppo = mpatches.Patch(label='Puppo', color = 'Blue')
floofer = mpatches.Patch(label='Floofer', color = 'Green')

plt.legend(handles=[puppers, doggo, puppo, floofer])
plt.title('Favorites v/s Retweets (Grouped by Dog Stages)')
plt.plot(x, line, color='gray', lw='1')
plt.grid(color='gray', linestyle='-', linewidth=1)

plt.show();
```

I plotted a scatterplot of all tweets from final dataframe.

It can also be observed that from the equation for the line of best fit that there tends to be 3 times as many favorites as retweets. ### COMMON BREEDS

```python
attributes = ['retweet_count', 'favorite_count', 'rating_denominator', 'rating_numerator']
df_dog_breeds = df_clean_final[attributes].copy()

def percentage_ratings(row):
    if row['rating_denominator'] == 0:
        return 0
    return row['rating_numerator']/row['rating_denominator']


df_dog_breeds['rating'] = df_dog_breeds.apply(percentage_ratings, axis=1)
```

```python
attributes = ['p1', 'p1_dog', 'p1_conf', 'p2', 'p2_dog', 'p2_conf', 'p3', 'p3_dog', 'p3_conf']
df_breed_possibility = df_clean_final[attributes].copy()

df_breed_possibility['breed'] = ''
df_breed_possibility['breed_conf'] = 1

def closest_breed(row):
    
    breed = 'not_dog'
    confidence = 0
    
    dog_preds = [(row[['p1', 'p1_dog', 'p1_conf']]), (row[['p2', 'p2_dog', 'p2_conf']]), (row[['p3', 'p3_dog', 'p3_conf']])]
    index_breed, index_isDog, index_conf = 0,1,2
    
    for pred in dog_preds:        
        if pred[index_isDog]:            
            if pred[index_conf] >= confidence:
                breed = pred[index_breed]
                confidence = pred[index_conf]
    
    # Updating the breed list
    row['breed'] = breed
    row['breed_conf'] = confidence
    return row
        
    breeds = [row['p1'], row['p2'], row['p3']]
    
df_breed_possibility = df_breed_possibility.apply(closest_breed, axis=1)

df_dog_breeds['breed'] = df_breed_possibility['breed']

df_dog_breeds['breed_conf'] = df_breed_possibility['breed_conf']
```
